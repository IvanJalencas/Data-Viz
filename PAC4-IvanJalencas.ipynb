```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('pax_all_agreements_data.xlsx')  
```

```python
df.head(5)
```

```python
df = df[["Con","Contp","Loc1ISO","PP","PPName","AgtId","Reg","Dat","Agtp","Stage","StDef","HrGen","ImUN","ImOth"]]
```

```python
df['Im3P'] = np.where((df['ImUN']==1)|(df['ImOth']==1), 1, 0)
```

```python
df
```

```python
idx = df.groupby(['PP'])['Dat'].transform(max) == df['Dat']
grouped = df[idx][['PP','Stage']]
grouped
```

```python
grouped['Ended'] = np.where(grouped['Stage'] == 'Cea', 1, 0)
```

```python
grouped
```

```python
grouped = grouped.drop_duplicates(subset=['PP', 'Ended'], keep='last')
```

```python
grouped
```

```python
merged = pd.merge(df, grouped, left_on='PP', right_on='PP')
```

```python
merged
```

```python
merged = merged.drop(['Stage_y'], axis=1)
```

```python
merged.to_excel("pax_ended_conflicts.xlsx")  
```
